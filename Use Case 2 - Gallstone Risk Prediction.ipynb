#Gallstone Risk Prediction Model using SQL

## What This Does
This SQL code predicts how likely each patient is to develop gallstones based on:
- Visceral fat (VFR)
- Total body fat (TBFR)
- BMI
- Gender

###Why these variables? (Based on Statistical Tests)

#####**Strongest Predictors (High Correlation + Significant p-values)**

**Visceral Fat Rating (VFR)**
- Correlation: 0.24 (highest)
- Statistically significant (p < 0.05)

**Total Body Fat Ratio (TBFR)**
- Correlation: 0.24
- Significant group differences

**BMI**
- Correlation: 0.18
- Statistically significant

**Gender**
- Chi-square test showed significant association (p < 0.05)
- Females at higher risk

The code then sorts patients into 3 risk groups:
- 🔴 **High Risk** (>70% chance)
- 🟡 **Medium Risk** (30-70% chance)
- 🟢 **Low Risk** (<30% chance)

## How It Works
1. **Input Data**: Uses patient health metrics
2. **Calculations**:
   - Applies medical research-based formulas
   - Converts raw scores to easy-to-understand percentages
3. **Output**:
   - Shows each patient's risk level
   - Flags who needs urgent care


```python
%sql
-- Step 1: Calculate proper coefficients directly from data statistics
WITH stats AS (
  SELECT
    -- Base rates
    COUNT(CASE WHEN `Gallstone Status` = 1 THEN 1 END) * 1.0 / COUNT(*) AS base_risk,
    
    -- Feature averages by class
    AVG(CASE WHEN `Gallstone Status` = 1 THEN `Visceral Fat Rating (VFR)` ELSE NULL END) AS vfr_mean_positive,
    AVG(CASE WHEN `Gallstone Status` = 0 THEN `Visceral Fat Rating (VFR)` ELSE NULL END) AS vfr_mean_negative,
    
    AVG(CASE WHEN `Gallstone Status` = 1 THEN `Total Body Fat Ratio (TBFR) (%)` ELSE NULL END) AS tbfr_mean_positive,
    AVG(CASE WHEN `Gallstone Status` = 0 THEN `Total Body Fat Ratio (TBFR) (%)` ELSE NULL END) AS tbfr_mean_negative,
    
    AVG(CASE WHEN `Gallstone Status` = 1 THEN `Body Mass Index (BMI)` ELSE NULL END) AS bmi_mean_positive,
    AVG(CASE WHEN `Gallstone Status` = 0 THEN `Body Mass Index (BMI)` ELSE NULL END) AS bmi_mean_negative,
    
    AVG(CASE WHEN `Gallstone Status` = 1 THEN Gender ELSE NULL END) AS gender_mean_positive,
    AVG(CASE WHEN `Gallstone Status` = 0 THEN Gender ELSE NULL END) AS gender_mean_negative
  FROM personal_catalog.gallstone_risk_prediction.gallstone_data
),

-- Step 2: Calculate normalized coefficients
coefficients AS (
  SELECT
    -- Intercept based on base risk
    LN(base_risk / (1 - base_risk)) AS intercept,
    
    -- Feature coefficients scaled by difference between positive/negative cases
    (vfr_mean_positive - vfr_mean_negative) / 10 AS vfr_coef,
    (tbfr_mean_positive - tbfr_mean_negative) / 10 AS tbfr_coef,
    (bmi_mean_positive - bmi_mean_negative) / 10 AS bmi_coef,
    (gender_mean_positive - gender_mean_negative) * 2 AS gender_coef
  FROM stats
)

-- Step 3: Apply to all patients with proper scaling
SELECT
  p.`Visceral Fat Rating (VFR)` AS VFR,
  p.`Total Body Fat Ratio (TBFR) (%)` AS TBFR,
  p.`Body Mass Index (BMI)` AS BMI,
  CASE WHEN p.Gender = 1 THEN 'Female' ELSE 'Male' END AS Gender,
  p.`Gallstone Status` AS actual_status,
  
  -- Scaled probability calculation
  ROUND(
    1 / (1 + EXP(-(
      c.intercept + 
      (p.`Visceral Fat Rating (VFR)` - vfr_mean_negative) * c.vfr_coef +
      (p.`Total Body Fat Ratio (TBFR) (%)` - tbfr_mean_negative) * c.tbfr_coef +
      (p.`Body Mass Index (BMI)` - bmi_mean_negative) * c.bmi_coef +
      (p.Gender - gender_mean_negative) * c.gender_coef
    ))) * 100, 
    1
  ) AS predicted_risk_percentage,
  
  -- Balanced risk categories (based on percentiles)
  CASE
    WHEN predicted_risk_percentage > 70 THEN '🔴High Risk'
    WHEN predicted_risk_percentage > 30 THEN '🟡Medium Risk'
    ELSE '🟢Low Risk'
  END AS risk_category
FROM personal_catalog.gallstone_risk_prediction.gallstone_data p
CROSS JOIN coefficients c
CROSS JOIN (SELECT vfr_mean_negative, tbfr_mean_negative, bmi_mean_negative, gender_mean_negative FROM stats) s
ORDER BY predicted_risk_percentage DESC;
```
